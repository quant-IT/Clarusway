```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=36
xbar=68
std=3
ci=0.90
```

```python
sem=std/np.sqrt(n)
sem
```

```python
stats.norm.interval(0.90, loc=xbar, scale=sem)
```
